## 4.5

```python
def is_prime(n):
    if n==2 or n==3: 
        return True
    if n%2==0 or n<2: 
        return False
    for i in range(3, int(n**0.5)+1, 2):  
        if n%i==0:
            return False    

    return True  


li = []
def print_prime(n):         # up to
    for i in range(2, n + 1):
        if is_prime(i):
            li.append(i)
    return li


print_prime(15)   
print(li)
```

```python
lis = []
def print_primes(n):        # first n
    i=2  
    cnt=0
    while(cnt<n):  
        if(is_prime(i)):  
            lis.append(i) 
            cnt+=1
        i+=1 
    return lis

print_primes(100)
print(lis)
```
