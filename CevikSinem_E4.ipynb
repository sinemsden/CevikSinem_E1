## 9.4

```python
n=200
def power_iter(Z,esp):
    u = np.ones(n)
    u = u.reshape(n,1)
    max_u = np.max(u)
    sub = np.inf
    iteration = 0

    while sub > esp:
        u = Z@u
        temp = np.max(u)
        sub = np.fabs(temp-max_u)
        max_u = temp
        u = u / max_u
        iteration += 1

    return max_u, iteration, u

esp = 1e-5
Z = np.random.normal(0,1,(n,n))
print("Max_eig, iterations, eig_vec")
print(power_iter(Z,esp))
```
