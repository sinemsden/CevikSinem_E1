## 10.1

```python
import numpy as np
import scipy
import scipy.optimize as optimize
from scipy.sparse.linalg import lsqr

def gmatrix(m, n): 
    if n >= m:
        raise ValueError('m should be greater than n')
    
    return np.random.rand(m,n)

def gvector(m): 
    return np.random.rand(m)
m, n = 10,4


x = gmatrix(5,2)
y = gvector(5)
a = optimize.nnls(x,y)


print('Norm of the residual ' + str(a[1]))
```

## 10.2

```python
def opt(x):
    return (np.power(np.sin(x-2), 2) * np.exp(-x**2))

def rev_opt(x):
    return -1 * opt(x)

rev_opt = optimize.minimize_scalar(rev_opt)
print(str(rev_opt.x)+"  "+str(opt(rev_opt.x)))
```

```python
from scipy.spatial.distance import pdist

n, m = 7, 3
X = np.random.rand(n, m)

distance = pdist(X)
print(distance)
```

# Matplotlib

## 11.1

```python
import matplotlib.pyplot as plt

def f(x):
    return (np.power(np.sin(x-2), 2) * np.exp(-x**2))


x = np.arange(0, 2, 0.00001, dtype = float)
plt.plot(x, f(x), 'r-')
plt.title("sin^2(x-2)e^(-(x^2))")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

## 11.2

```python
b = np.random.rand(10)
X = np.random.rand(20, 10)
z = np.random.standard_normal(size=20)
y = np.dot(X,b) + z

b_estimator = optimize.nnls(X,y)[0]

plt.plot(b,b_estimator)
plt.xlabel('real value')
plt.ylabel('estimation')

plt.show()
```

```python
from scipy import stats

z = np.random.rayleigh(10, 10000)
dens = stats.kde.gaussian_kde(z)
foo = np.linspace(-2, 10, 100)

plt.hist(dens(foo), bins=25, color="green")
plt.show()
```
